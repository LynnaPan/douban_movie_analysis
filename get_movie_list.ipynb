```python
from urllib import request
res = request.urlopen("https://movie.douban.com/cinema/nowplaying/shanghai/")
html_data = res.read().decode('utf-8')
#print(html_data)
```

```python
from bs4 import BeautifulSoup as bs
soup = bs(html_data, "html.parser")
nowplaying = soup.find_all('div', id='nowplaying')
nowplaying_list = nowplaying[0].find_all('li', class_='list-item')
print(nowplaying_list)
```

```python
nowplaying = []
for item in nowplaying_list:
    one_movie = {}
    one_movie['id'] = item['id']
    one_movie['name'] = item.img['alt']
    nowplaying.append(one_movie)
print(nowplaying)
```

```python
comment_url = "https://movie.douban.com/subject/"+nowplaying[0]['id']+"/comments?status=P"
res = request.urlopen(comment_url)
html_data = res.read().decode('utf-8')
soup = bs(html_data, "html.parser")
comments = soup.find_all('div', class_ = "comment")
comments_content = ""
for one_comment in comments:
    cleaned_comment = str(one_comment.p.string).strip()
    comments_content = comments_content + cleaned_comment

# 清除标点符号
import re
pattern = re.compile(r'[\u4e00-\u9fa5]+')
filterdata = re.findall(pattern, comments_content)
cleaned_comments = ''.join(filterdata)
print(cleaned_comments)
```

```python
import jieba
import pandas as pd


words = jieba.lcut(cleaned_comments)
stopwords=pd.read_csv("stopwords.txt",index_col=False,quoting=3,sep="\t",names=['stopword'], encoding='utf-8')#quoting=3全不引用
words_df=pd.DataFrame({'words':words})
words_df=words_df[~words_df.words.isin(stopwords.stopword)]
print(words_df)

```

```python
import numpy    #numpy计算包
words_stat=words_df.groupby(by=['words'])['words'].agg({"count":numpy.size})
words_stat=words_stat.reset_index().sort_values(by=["count"],ascending=False)
print(words_stat)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
matplotlib.rcParams['figure.figsize']=(10.0, 5.0)
from wordcloud import WordCloud



wordcloud=WordCloud(font_path="simhei.ttf",background_color="black",max_font_size=80) #指定字体类型、字体大小和字体颜色
word_frequence = {x[0]:x[1] for x in words_stat.head(1000).values}
word_frequence_list = []
for key in word_frequence:
    temp = (key,word_frequence[key])
    word_frequence_list.append(temp)
 
wordcloud=wordcloud.fit_words(word_frequence_list)
plt.imshow(wordcloud)
```
